```python
y = df['Survived']
```

```python
df.drop(columns=['Survived'], axis=1, inplace=True)
```

```python
df_encoded = pd.get_dummies(df, columns=df.select_dtypes('object').columns, dtype='float')
```

```python
len(df_encoded.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(df_encoded)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
kNeighbors = KNeighborsClassifier(n_neighbors=5)

kNeighbors.fit(X_train, y_train)

y_pred_knn = kNeighbors.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(y_test, y_pred_knn)
```

```python
print(classification_report(y_test, y_pred_knn))
```

```python
print(confusion_matrix(y_test, y_pred_knn))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_knn), annot=True)
```

```python
def show_score(y_real, y_pred, model_name):
    print(f'{model_name} Accuracy Score: ')
    print(accuracy_score(y_real, y_pred))
    print((f'{model_name} Classification Report: '))
    print(classification_report(y_real, y_pred))
    print((f'{model_name} Confusion Matrix: '))
    print(confusion_matrix(y_real, y_pred))
```

```python
show_score(y_test, y_pred_knn, "KNN Classifier")
```

```python
clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

sklearn.tree.plot_tree(clf)
```

```python
y_pred_decision_tree = clf.predict(X_test)
```

```python
show_score(y_test, y_pred_decisionTree, "Decision Tree")
```

```python
svm = SVC()

svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

show_score(y_test, y_pred_svm, "Support Vector Machines")
```

```python
log_reg = LogisticRegression()

log_reg.fit(X_train, y_train)

y_pred_log_reg = log_reg.predict(X_test)

show_score(y_test, y_pred_log_reg, "Logistic Regression")
```

```python
naive_bayes = GaussianNB()

naive_bayes.fit(X_train, y_train)

y_pred_naive_bayes = naive_bayes.predict(X_test)

show_score(y_test, y_pred_naive_bayes, "Naive Bayes")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search_knn = GridSearchCV(estimator=KNeighborsClassifier(), 
                               param_grid={'n_neighbors': [1, 2, 4, 5, 6, 8, 10, 12, 15, 20, 24, 28, 30],
                                           'weights': ['uniform', 'distance'],
                                           'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']},
                               scoring='accuracy',
                               n_jobs=10,
                               refit=False,
                               cv=10
                              )

grid_search_knn.fit(X, y)

grid_search_knn.best_params_
grid_search_knn.best_score_
grid_search_knn.cv_results_ ## turn into DATAFRAME
```

```python
def grid_search_best(grid_search):
    print('Best parameters found: ', grid_search.best_params_)
    print('Best Score: ', grid_search.best_score_)
    cv_res_df = pd.DataFrame(grid_search.cv_results_)
    print("CV Overall results: ")
    print(cv_res_df)
```

```python
grid_search_best(grid_search_knn)
```

```python
grid_search_des_tree = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid={
        'criterion': ["gini", "entropy", "log_loss"],
        'max_depth': [1, 2, 4, 6, 8, 10, 12, 14, 18, 20, 24, 28, 32, 64, 128, 256],
        'splitter': ["best", "random"]
    },
    scoring='accuracy',
    n_jobs=10,
    cv=10,
    refit=False)

grid_search_des_tree.fit(X, y)
```

```python
grid_search_best(grid_search_des_tree)
```
